```python
def length_lcs(a, b):
    m, n = len(a), len(b)
    # Inizializziamo la tabella direttamente con 0
    memo = [[0 for _ in range(n + 1)] for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if a[i-1] == b[j-1]:
                # Se i caratteri sono uguali, prendi la diagonale + 1
                memo[i][j] = 1 + memo[i-1][j-1]
            else:
                # Altrimenti prendi il massimo tra sopra e sinistra
                memo[i][j] = max(memo[i-1][j], memo[i][j-1])

    return memo[m][n]
```

```python
def reconstruct_lcs(a,b):
  risultato = [] # stringa che alla fine conterrà la LCS
  m = len(a)
  n = len(b)
  memo = [[0 for _ in range(n + 1)] for _ in range(m + 1)] # creo la tabella di memoization

  for i in range(1,m+1):
    for j in range(1,n+1):
      if a[i-1] == b[j-1]:
        memo[i][j] = 1 + memo[i-1][j-1]
      else:
        memo[i][j] = max(memo[i-1][j], memo[i][j-1])

  i = m
  j = n
  while i > 0 and j > 0:   # backtracking
    if a[i-1] == b[j-1]:
      risultato.append(a[i-1])
      i -= 1
      j -= 1
    elif memo[i-1][j] > memo[i][j-1]:
      i -= 1
    else:
      j -= 1

  risultato = risultato[::-1]
  return risultato
```

```python
# Test 2: Sequenze complesse
dna_1 = "ACCGGTCGAGTGCGCGGAAGCCGGCCGAA"
dna_2 = "GTCGTTCGGAATGCCGTTGCTCTGTAAA"

# Risultato atteso:
# LCS suggerita: "GTCGTCGGAAGCCGGCCGAA" (o simili, la lunghezza dovrebbe essere circa 20)
print("la lunghezza della LCS è: ",length_lcs(dna_1,dna_2))
print("la LCS è: ",reconstruct_lcs(dna_1,dna_2))
```
